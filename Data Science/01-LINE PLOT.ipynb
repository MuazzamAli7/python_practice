# Adding Titles

```python
#STEP 1....IMPORT LIBRARIES
import seaborn as sns
import matplotlib.pyplot as plt
#STEP 2....LOAD DATA SET
flower=sns.load_dataset('iris') #iris is a data in seaborn lib
#step 3...draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=flower)
plt.title('variation of sepal L wrt Sepal W')
plt.show()
```

# Adding limits of X & Y AXES

```python
#STEP 1....IMPORT LIBRARIES
import seaborn as sns
import matplotlib.pyplot as plt
#STEP 2....LOAD DATA SET
flower=sns.load_dataset('iris') #iris is a data in seaborn lib
#step 3...draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=flower)
plt.title('variation of sepal L wrt Sepal W')
plt.xlim(1)
plt.ylim(1)
plt.show()
```

# SIZE OF PLOT

```python
#STEP 1....IMPORT LIBRARIES
import seaborn as sns
import matplotlib.pyplot as plt
#STEP 2....LOAD DATA SET
flower=sns.load_dataset('iris') #iris is a data in seaborn lib
#step 3...draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=flower)
plt.title('variation of sepal L wrt Sepal W')
plt.xlim(1)
plt.ylim(1)
plt.figure(figsize=(18,20))
plt.show()
```
